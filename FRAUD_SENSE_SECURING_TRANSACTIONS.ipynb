### IQR TEST:

```python
def Iqr(col):
    Q1=credit_df[col].quantile(0.25)
    Q3=credit_df[col].quantile(0.75)
    IQR=Q3-Q1
    upper_limit=Q3+1.5*IQR
    lower_limit=Q1-1.5*IQR
    credit_df[col]=np.where(credit_df[col]>upper_limit,upper_limit,np.where(credit_df[col]<lower_limit,lower_limit,credit_df[col]))
    
    
for i in list(credit_df.columns)[1:29]:
    Iqr(i)
    
```

## Model Building

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.naive_bayes import GaussianNB,BernoulliNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import cross_val_score
```

```python
#logistic
logistic=LogisticRegression()
```

```python
#DecisionTree
dtree=DecisionTreeClassifier()
```

```python
#Randomforest
rf=RandomForestClassifier()
```

```python
#Boosting
gdb=GradientBoostingClassifier()
xgb=XGBClassifier()
```

```python
#NAIVEBAYES
gnb=GaussianNB()
bnb=BernoulliNB()
```

```python
#K nearest neighbor
knn=KNeighborsClassifier()
```

```python
##GRADIENT DESCENT
GDC_model=SGDClassifier()
```

```python
list1=[(logistic,'logistic',1),(dtree,'decisiontree',2),(rf,'randomforest',3),(gdb,'GradientBoost',4),(xgb,'XGBoost',5),(gnb,'GradientNB',6),(bnb,'BernoulliNB',7),(GDC_model,'GDC',8)]
```

```python
for model,model_name,i in list1:
    model.fit(x_train,y_train)
    y_pred_train=model.predict(x_train)
    y_pred_test=model.predict(x_test)
    print('===================================================================')
    print(f'Cross validation of {model_name}')
    Train_acc=cross_val_score(model,x_train,y_train,cv=3).mean()
    Test_acc=cross_val_score(model,x_test,y_test,cv=3).mean()
    varience=abs(Train_acc-Test_acc)
    print('Train accuracy:',Train_acc)
    print('Test accuracy:',Test_acc)
    print('varience:',varience)
    if varience>10 :
        print(f'Overfitting {model_name}')   
```

```python
#classification report
from sklearn.metrics import classification_report
for model,model_name,i in list1:
    model.fit(x_train,y_train)
    y_pred_train=model.predict(x_train)
    y_pred_test=model.predict(x_test)
    print('===================================================================')
    print(f'Classification report of {model_name}')
    print('====================================================================')
    Train_cvr=classification_report(y_train,y_pred_train)
    Test_cvr=classification_report(y_test,y_pred_test)
    print('Train classification report:',Train_cvr)
    print('===================================================================')
    print('Test classification report:',Test_cvr)
```

```python
#Confusion matrix:
from sklearn.metrics import confusion_matrix
for model,model_name,i in list1:
    model.fit(x_train,y_train)
    y_pred_train=model.predict(x_train)
    y_pred_test=model.predict(x_test)
    cm=confusion_matrix(y_test,y_pred_test)
    print(f'Confusion matrix of {model_name}')
    plt.figure(figsize=(5,5))
    plt.subplot(5,2,i)
    sns.heatmap(cm,annot=True,cmap='coolwarm')
    plt.show()
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
```

```python
for model,model_name,i in list1:
    model.fit(x_train,y_train)
    y_pred_train=model.predict(x_train)
    y_pred_test=model.predict(x_test)
    roc_auc=roc_auc_score(y_test,y_pred_test)
    fpr,tpr,thresholds=roc_curve(y_test,y_pred_test)
    plt.plot(fpr,tpr,label=f'ROC CURVE{model_name}(area=%0.2f)'%roc_auc)
    plt.plot([0,1],[0,1],'k--')
    plt.xlim([0.0,1.0])
    plt.ylim([0.0,1.0])
    plt.xlabel('FPR')
    plt.ylabel('TPR')
    plt.table('ROC WITH AUC')
    plt.legend(loc='lower right')
plt.show()
```
